## Exercício 17 — Análise de mapa de calibração de motor

Mapas de calibração relacionam RPM × Carga do motor a parâmetros como avanço de ignição, injeção de combustível, etc. Vamos simular um cenário simplificado:


````python
# Simulando dados de um mapa de calibração
rpm_vals = np.arange(1000, 7000, 500)
carga_vals = np.arange(10, 110, 10)  # percentual de carga

rpm_grid, carga_grid = np.meshgrid(rpm_vals, carga_vals)
# Simulando avanço de ignição (graus antes do PMS)
avanco_ignição = 35 - 0.003 * rpm_grid - 0.15 * carga_grid + 0.000002 * rpm_grid * carga_grid
# Adicionando um pouco de ruído
avanco_ignição += np.random.normal(0, 0.5, avanco_ignição.shape)

df_mapa = pd.DataFrame({
    'rpm': rpm_grid.flatten(),
    'carga_pct': carga_grid.flatten(),
    'avanco_ignicao_graus': avanco_ignição.flatten().round(1)
})
print(df_mapa.head(10))
print(f"\nShape: {df_mapa.shape}")
````

**Tarefas:**
1. Crie um gráfico 3D ou heatmap que represente o **mapa de avanço de ignição** (RPM no eixo X, Carga no eixo Y, avanço como cor).

3. Calcule o avanço médio para cada faixa de RPM (agrupe a cada 1000 RPM) e plote como barras.
4. Identifique a combinação RPM × Carga com o **menor** avanço de ignição.


````python
# SEU CÓDIGO AQUI

````

<details>
<summary>💡 Dica para heatmap do mapa de calibração</summary>
